```python
!pip install transformers torch gradio pillow

import gradio as gr
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image

processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-small")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-small")

def generate_caption(image):
    image = image.resize((384, 384))  # resize for CPU efficiency
    inputs = processor(images=image, return_tensors="pt")
    outputs = model.generate(**inputs, num_beams=3, max_length=50)
    caption = processor.decode(outputs[0], skip_special_tokens=True)
    return caption

def caption_image(image):
    try:
        return generate_caption(image)
    except Exception as e:
        return f"An error occurred: {e}"

iface = gr.Interface(
    fn=caption_image,
    inputs=gr.Image(type="pil"),
    outputs="text",
    title="Image Captioning with BLIP",
    description="Upload an image to generate captions"
)

iface.launch(server_name="127.0.0.1", server_port=7861)
```
